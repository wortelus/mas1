```python
import numpy as np
import random
```

```python
adj_list_type = [set[int]]
```

```python
def select_nodes(adj_list: adj_list_type, m: int):
    weights = [len(neighbors) for neighbors in adj_list]
    nodes = list(range(len(adj_list)))
    sampled_nodes = random.choices(nodes, weights=weights, k=m)
    
    return sampled_nodes


def ba_model_append(adj_list: adj_list_type, m: int):
    connections = select_nodes(adj_list, m)
    # for i in connections:
    #     adj_list[i].add(len(adj_list))
    #     adj_list.append({i})

    connections = select_nodes(adj_list, m)

    # Přidat nový uzel (s prázdným seznamem hran)
    new_node_idx = len(adj_list)
    adj_list.append(set())
    
    # Přidat hrany mezi novým uzlem a vybranými uzly
    for i in connections:
        adj_list[new_node_idx].add(i)
        adj_list[i].add(new_node_idx)


def ba_model(n0_adjacency_list: adj_list_type, m: int, t: int, max_nodes: int = None):
    # also dynamically calculated as 2mt
    degree_sum = np.sum([len(a) for a in n0_adjacency_list])
    
    adj_list = n0_adjacency_list[:]
    for i in range(t):
        print(f"Before Step {i}, degree sum: {degree_sum}, number of nodes: {len(adj_list)}")
        ba_model_append(adj_list, m)
        degree_sum += m
        
        if max_nodes is not None and len(adj_list) >= max_nodes:
            break
    
    return adj_list    
```

```python
# taken from cv6
def export_edge_list(graph, filename):
    with open(filename, "w") as f:
        f.write("source;target\n")
        for node, neighbors in enumerate(graph):
            for neighbor in neighbors:
                f.write(f"{node};{neighbor}\n")
```

```python
# n0_adjacency_list = [{1, 2}, {0, 2}, {0, 1, 3}, {2}]
n0_adjacency_list = [{1, 2}, {0, 2}, {0, 1}]
m = 3
t = 100000
max_nodes = 550
adj_list = ba_model(n0_adjacency_list, m, t, max_nodes=max_nodes)

export_edge_list(adj_list, f"ba_model_m{m}.csv")
```
